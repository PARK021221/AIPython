## 누적합/누적곱 함수로 만들어보기

````python
total = 0
fac = 1
for i in range(1, 6):
    total = total + i
    fac = fac * i
print("sum(1..5):", total)
print("fac(1..5):", fac)
````

````python
def sum(start, end) :
  total=0
  for i in range(start, end+1):
    total = total + i
  return total
def fac(start, end) :
  fac = 1
  for i in range(start, end+1):
    fac = fac * i
  return fac

num1=int(input("정수1 입력:"))
num2=int(input("정수2 입력:"))
print(f"sum({num1}..{num2}):", sum(num1,num2))
print(f"fac({num1}..{num2}):", fac(num1,num2))
````

#딕셔너리
* {}에 키-값 쌍(key-value 구조)으로 구성된 여러개의 아이템을 ,로 구분하여저장
* 각 아이템의 키를 활용하여 값의 선택, 변경, 추가, 삭제 가능

*  get() 사용 권장


## 텍스트 파일 읽기: read / readline / readlines


1. read(): 파일 전체를 한 번에 읽기


````python
path = "/content/prompt_CheatSheet.txt"

with open(path, "r", encoding="utf-8") as f:
    content = f.read()

print(content)
````

2. readline(): 한 줄씩 읽기(첫 줄만 예시)


````python
with open(path, "r", encoding="utf-8") as f:
    line1 = f.readline()
    line2 = f.readline()

print("1줄:", line1.strip())
print("2줄:", line2.strip())
````

3. readline(): 한 줄씩 읽기(while 문을 사용하여 모든 줄 읽기)


````python
with open(path, "r", encoding="utf-8") as f:
    print("파일의 모든 줄을 읽습니다:\n")
    while True:
        line = f.readline()
        if not line:  # 빈 문자열이면 파일의 끝에 도달했으므로 루프 종료
            break
        print(line.strip()) # 각 줄의 양쪽 공백(줄바꿈 포함) 제거 후 출력
````

4. readlines(): 줄 단위 리스트로 읽기


````python
with open(path, "r", encoding="utf-8") as f:
    lines = f.readlines()

print("총 줄 수:", len(lines))
print("마지막 줄:", lines[-1].strip())
````

## grade.csv 성적 데이터 읽고 평균 구하기


### 3-1. csv 모듈로 읽기


````python
import csv

path_csv = "/content/drive/MyDrive/Colab Notebooks/0628-0701강의/grade.csv"

rows = []
with open(path_csv, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        rows.append(row)

print("행 개수:", len(rows))
print("첫 행:", rows[0])
````

### 3-2. 각 학생의 평균 구하기


````python
def to_int(x):
    return int(x)

for row in rows:
    name = row["name"]
    korean = to_int(row["korean"])
    math = to_int(row["math"])
    english = to_int(row["english"])
    avg = (korean + math + english) / 3
    print(f"{name}: avg={avg:.2f}")
````
