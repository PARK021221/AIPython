## 5. 거스름돈 계산기


역할: 너는 파이썬 튜터야

목표: 아래 문제를 파이썬으로 풀어줘.

문제: 거스름돈 계산기를 만들어줘. 사용자가 1만원 단위의 금액을 입력하면, 상품값을 제외한 거스름돈은 5천원, 1천원, 500원, 100원의 거스름돈을 받을 수 있도록 해줘.

입력: (예) 지불금액  : 20000 상품금액 : 15400

출력: (예)==거스름돈 ==
5,000원권 : 0장 1,000원권 : 4장 500원 동전 : 0개  100원 동전 : 3개 (각 권종당 줄바꿈으로 처리)

제약: 몫과 나머지 연산자 사용, 1000원 이상은 지폐 500원, 100원은 동전으로 처리, 사용자가 10,000원 단위로 입력할 수 있도록 제시.

요청: 코드에 대해 한 줄씩 설명하고 테스트 3개 실행


```python
# TODO: 거스름돈 계산기
지불금액 = int(input("지불금액 (10,000원 단위): "))
상품금액 = int(input("상품금액: "))

change = 지불금액 - 상품금액

# 각 화폐 단위별 개수 계산
n5000 = change // 5000
rem = change % 5000

n1000 = rem // 1000
rem = rem % 1000

c500 = rem // 500
rem = rem % 500

c100 = rem // 100

print(f"\n==거스름돈: {change:,.0f}원 ==")
print(f"5,000원권 : {n5000}장")
print(f"1,000원권 : {n1000}장")
print(f"500원 동전 : {c500}개")
print(f"100원 동전 : {c100}개")
```

## 8. 성적 분석 프로그램 만들기


### 문제 : 점수 5 개를 입력받아 리스트로 저장
출력: 평균/최대/최소/합계

도전: 90점 이상 개수 카운트하여 출력

"테스트 입력”을 함께 만들기


```python
# TODO: 성적 분석 프로그램 만들기 (for 문 사용)

scores = []

# 1. 5번 입력 받기
for i in range(5):
    s = int(input(f"{i + 1}번째 점수 입력: "))
    scores.append(s)

# 2. 초기값 설정
total_sum = 0
max_score = scores[0]
min_score = scores[0]
count_90 = 0

# 3. for 문으로 합계, 최대, 최소, 90점 이상 개수 계산
for s in scores:
    total_sum += s

    if s > max_score:
        max_score = s

    if s < min_score:
        min_score = s

    if s >= 90:
        count_90 += 1

avg = total_sum / len(scores)

# 4. 결과 출력
print(f"\n--- 분석 결과 ---")
print(f"입력된 점수: {scores}")
print(f"합계: {total_sum}")
print(f"평균: {avg:.1f}")
print(f"최대값: {max_score}")
print(f"최소값: {min_score}")
print(f"90점 이상 개수: {count_90}개")

# 테스트 입력 예시: 85, 92, 78, 100, 65
```

## 9. ToDo 리스트 만들기


### 문제 : 리스트에 할 일을 저장
add: 입력받아 append

list: 현재 목록 출력(번호 포함)

done: 번호를 입력받아 pop/remove로 삭제


```python
# TODO: TODO 리스트 만들기 (add, list, done 기능 구현)
todo_list = []

while True:
    print("\n--- To-Do List Menu ---")
    print("1. add (할 일 추가)")
    print("2. list (목록 출력)")
    print("3. done (할 일 삭제)")
    print("4. exit (종료)")

    menu = input("메뉴를 선택하세요: ")

    if menu == '1':
        # add: 입력받아 append
        task = input("추가할 할 일을 입력하세요: ")
        todo_list.append(task)
        print(f"'{task}' 추가 완료.")

    elif menu == '2':
        # list: 현재 목록 출력(번호 포함)
        print("\n[현재 할 일 목록]")
        if not todo_list:
            print("할 일이 없습니다.")
        else:
            for i in range(len(todo_list)):
                print(f"{i + 1}. {todo_list[i]}")

    elif menu == '3':
        # done: 번호를 입력받아 pop으로 삭제
        if not todo_list:
            print("삭제할 항목이 없습니다.")
        else:
            num = int(input("완료(삭제)할 번호를 입력하세요: "))
            if 1 <= num <= len(todo_list):
                removed = todo_list.pop(num - 1)
                print(f"{num}번 '{removed}' 삭제 완료.")
            else:
                print("잘못된 번호입니다.")

    elif menu == '4':
        print("프로그램을 종료합니다.")
        break

    else:
        print("1~4 사이의 번호를 입력해주세요.")
```

```python
# TODO: TODO 리스트 만들기 (예외 처리 추가: 빈 리스트 및 인덱스 체크)
todo_list = []

while True:
    print("\n--- To-Do List Menu ---")
    print("1. add (할 일 추가)")
    print("2. list (목록 출력)")
    print("3. done (할 일 삭제)")
    print("4. exit (종료)")

    menu = input("메뉴를 선택하세요: ")

    if menu == '1':
        task = input("추가할 할 일을 입력하세요: ")
        todo_list.append(task)
        print(f"'{task}' 추가 완료.")

    elif menu == '2':
        print("\n[현재 할 일 목록]")
        # 빈 리스트 처리
        if not todo_list:
            print("목록이 비어있습니다. 할 일을 먼저 추가해주세요.")
        else:
            for i in range(len(todo_list)):
                print(f"{i + 1}. {todo_list[i]}")

    elif menu == '3':
        # 빈 리스트 처리
        if not todo_list:
            print("삭제할 항목이 없습니다.")
        else:
            try:
                user_input = input("완료(삭제)할 번호를 입력하세요: ")
                num = int(user_input)

                # 인덱스 범위 체크
                if 1 <= num <= len(todo_list):
                    removed = todo_list.pop(num - 1)
                    print(f"{num}번 '{removed}' 삭제 완료.")
                else:
                    print(f"입력 오류: 1부터 {len(todo_list)} 사이의 번호를 입력하세요.")
            except ValueError:
                print("오류: 숫자를 입력해야 합니다.")

    elif menu == '4':
        print("프로그램을 종료합니다.")
        break

    else:
        print("1~4 사이의 번호를 입력해주세요.")
```


## 10. 숫자 맞히기 게임(while)


### 요구사항
- 정답 숫자를 맞힐 때까지 반복 입력
- 힌트: 작으면 UP, 크면 DOWN
- 추가: 시도 횟수 카운트
- 무한 루프 방지: 시도 횟수 제한(max_tries) 같은 안전장치를 두어도 좋습니다.


```python
import random

# 정답 숫자 생성 (1~50 사이의 랜덤 숫자)
target = random.randint(1, 50)
tries = 0
max_tries = 10

print(f"1부터 50 사이의 숫자를 맞혀보세요! (기회: {max_tries}번)")

while tries < max_tries:
    try:
        guess = int(input(f"{tries + 1}번째 시도: "))
        tries += 1

        if guess < target:
            print("UP!")
        elif guess > target:
            print("DOWN!")
        else:
            print(f"축하합니다! {tries}번 만에 정답({target})을 맞히셨습니다.")
            break
    except ValueError:
        print("숫자만 입력 가능합니다.")

if tries == max_tries and guess != target:
    print(f"아쉽네요. 모든 기회를 사용하셨습니다. 정답은 {target}이었습니다.")
```

### 01.  선박 제원 출력 및 배수량 계산 [산술연산자]

문제: 사용자의 입력을 받아 선박의 이름, 길이(L), 폭(B), 흘수(T)를 변수에 저장하고, 단순화된 공식을 이용해 배수량(∆)을 계산하여 출력하는 프로그램을 작성하세요.
공식: Displacement = L * B * T * Cb * 1.025 (단, 방형비 Cb 는 0.8로 가정)

```python
# 01. 선박 제원 출력 및 배수량 계산

# 사용자 입력 받기
ship_name = input("선박 이름을 입력하세요: ")
L = float(input("길이(L)를 입력하세요 (m): "))
B = float(input("폭(B)을 입력하세요 (m): "))
T = float(input("흘수(T)를 입력하세요 (m): "))

# 변수 설정
Cb = 0.8
Density = 1.025

# 배수량 계산 (산술 연산자 사용)
displacement = L * B * T * Cb * Density

# 결과 출력
print(f"\n--- {ship_name} 분석 결과 ---")
print(f"계산된 배수량: {displacement:.2f} tons")
```

### 02. 선체 구조 부재 무게 등급 판별   [제어문]

문제: 선박에 사용되는 강판의 무게를 입력받아, 무게에 따라 크레인 운송 등급(A, B, C)을 판별하는 프로그램을 작성하세요.
조건: 50톤 이상은 'A등급', 20톤 이상 50톤 미만은 'B등급', 20톤 미만은 'C등급'으로 분류합니다.

```python
# 02. 선체 구조 부재 무게 등급 판별

# 무게 입력 받기
weight = float(input("강판의 무게를 입력하세요 (tons): "))

# 제어문을 통한 등급 판별
if weight >= 50:
    grade = "A등급"
elif weight >= 20:
    grade = "B등급"
else:
    grade = "C등급"

# 결과 출력
print(f"\n입력 무게: {weight} tons")
print(f"판별 결과: {grade} (크레인 운송 대상)")
```

### 03.  강재 리스트 반복 출력 및 합계 [반복문]

문제: 리스트에 저장된 여러 개의 강재 무게 데이터를 반복문을 통해 하나씩 출력하고, 전체 강재의 총 무게를 계산하는 프로그램을 작성하세요.

```python
# 03. 강재 리스트 반복 출력 및 합계

# 강재 무게 데이터 리스트
steel_weights = [15.5, 22.0, 8.4, 33.2, 12.7]

total_weight = 0

print("[강재별 무게 출력]")
# 반복문을 통해 데이터 출력 및 합계 계산
for weight in steel_weights:
    print(f"강재 무게: {weight} tons")
    total_weight += weight

# 최종 합계 출력
print(f"\n전체 강재 총 무게: {total_weight:.2f} tons")
```
